```python
covid2_dt_deathCnt = covid2.groupby('createDt').deathCnt.mean()
covid2_dt_deathCnt
```

```python
covid2_dt_deathCnt_df = covid2_dt_deathCnt.to_frame()
covid2_dt_deathCnt_df
```

```python
covid2_dt_deathCnt_df.reset_index(inplace=True)
```

```python
covid2_dt_deathCnt_df.columns
```

```python
covid2_dt_deathCnt_df['createDt'] = pd.to_datetime(covid2_dt_deathCnt_df['createDt'])
```

```python
covid2_dt_deathCnt_df.dtypes
```

```python
covid2_dt_deathCnt_df['createDt_cat'] = pd.cut(covid2_dt_deathCnt_df['createDt'],
                        bins = [pd.to_datetime('2020-1-1'), pd.to_datetime('2020-7-1'), pd.to_datetime('2020-12-31')],
                        include_lowest = True,
                        labels = ['pre_year', 'post_year']         
                        )
```

```python
covid2_dt_deathCnt_df
```

```python
covid2_dt_deathCnt_df_count = covid2_dt_deathCnt_df.groupby('createDt_cat').count()
```

```python
covid2_dt_deathCnt_df_mean = covid2_dt_deathCnt_df.groupby('createDt_cat').mean()
```

```python
covid2_dt_deathCnt_df_sum = covid2_dt_deathCnt_df.groupby('createDt_cat').sum()
```

```python
covid2_dt_deathCnt_df_count['label'] = ['pre','post']
```

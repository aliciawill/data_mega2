```python
# enumerate 반복문: index, 반복할 item을 둘 다 꺼내주고 싶을 때 사용
data = [100, 200, 300]

for i, item in enumerate(data, 0):  # 0은 index의 start값
    print(i, ':', item)
```

```python
# enumerate(열거하다) 함수: for each 처럼 하나씩 꺼내올 수 있음
# Jquery의 $('list').each: index랑 list의 항목을 하나씩 꺼냄
# item에는 시리즈의 index와 values를 tuple로 반복해서 꺼내온다.

for i, item in enumerate(item_count.iteritems(), 1):  # iterate: 반복하다, 1: index의 start값
    print('TOP', i, ':', item)
```

```python
for i, item in enumerate(item_count.iteritems(), 0):  # iterate: 반복하다, 1개씩 꺼내라
    print('TOP', i+1, ':', item[0], item[1])
```

```python
for i, (pre, post) in enumerate(item_count.iteritems(), 0):  # iterate: 반복하다, 1개씩 꺼내라
    print('TOP', i+1, ':', pre, post)
```

```python
# 시리즈의 values만 꺼내오고 싶으면, 시리즈를 enumerate에 입력값으로 넣어주면 됨
# 시리즈를 tolist() 하면 시리즈의 values만 리스트로 만들어진다.

for i, item in enumerate(item_count.tolist(), 0):  # 0: index의 start 값
    print('TOP', i + 1, ':', item)  # values만 꺼내올 수 있다
```

```python
# 시리즈의 모든 행에 있는 string을 추출해서 $ 빼야한다.(인덱싱을 사용하기로 함)
# 시리즈를 모두 변경하기 위해서는 for문 있어야 함.
# float()함수는 비파괴 => float() 캐스팅한 결과를 리스트에 모아놓았다가,
# 반복문이 모두 끝나면 리스트에 모아놓은 데이터를 df의 컬럼에 다시 집어넣는다.
```

```python
# price 한개씩 가져와서 $ 사인을 빼고 숫자만 가져온 후, float로 캐스팅 해줌
result = []
for x in food['item_price']:  
    result.append(float(x[1:]))
result[:4620]
```

```python
# 반복문이 모두 끝나면 리스트에 모아놓은 데이터를 df의 컬럼에 다시 집어넣는다.
food['item_price'] = result
```

# <Step4. 탐색적 분석> : 스무고개로 분석하는 개념적 탐색
#### 데이터를 이해하기 위한 조금 더 복잡한 질문들로 탐색적 데이터 분석 연습하기
#### 주문당 평균 계산금액 출력하기
#### 한 주문에 40달러 이상 사용한 주문의 id들 출력하기
#### 각 아이템의 가격 구하기
#### 가장 비싼 주문에서 item이 몇개 팔렸는지 구하기
#### "Veggie Salad Bowl"이 몇 번 주문되었는지 구하기
#### "Chicken Bowl"을 2개 이상 주문한 주문 횟수 구하기

```python
# 주문당 평균 계산금액 출력
df_order = food.groupby('order_id')['item_price'].sum()  # order_id를 기준으로 그룹화하여 합계를 구한 후, 평균 구하기
df_order.mean()
```

```python
# 한 주문에 40달러 이상 사용한 주문의 id들 출력하기
# for문으로 id당 주문금액 합계가 40이상일 경우 
# result2 리스트에 값을 넣어주기
result2 = []
for x in range(len(df_order)):
    if(list(df_order)[x] >= 40):
        result2.append(x)        
result2[:10]
```

```python
# 각 아이템의 가격 구하기
# item_name을 기준으로 그룹화하여 item_price의 최소값 구하기
df_item = food.groupby('item_name')['item_price'].min()
df_item[:10]
```

```python
# 가장 비싼 주문에서 item이 몇개 팔렸는지 구하기
# food를 item_price를 기준으로 내림차순으로 정렬시키고 
# df_order[df_order['item_price']== 205.25]
maxValue = df_order.max()
maxValue
```

```python
# "Veggie Salad Bowl"이 몇 번 주문되었는지 구하기
# item_name중 메뉴명이 동일한 것을 카운트하고 quantity에 맞는 값을 찾기
food[food['item_name'] == "Veggie Salad Bowl"].count()['quantity']
```

```python
# "Chicken Bowl"을 2개 이상 주문한 주문 횟수 구하기
# 메뉴에 해당하는 것 중 quantity(주문)이 2회 이상인 것들의 len 가져오기
CB = food[food['item_name'] == "Chicken Bowl"]
len(CB[CB['quantity'] >= 2])
```
